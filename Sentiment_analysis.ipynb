```python
lowercase=text.lower()#.lower() is a built-in Python method primarily used for string handling.
cleanedtext=lowercase.translate(str.maketrans('','',string.punctuation))#To remove punctuation from a string in Python, you can use the str. translate() method with the . maketrans() method.
tokenizedwords=word_tokenize(cleanedtext,"english")#NLTK provides a function called word_tokenize() for splitting strings into tokens (nominally words).
```

#Categorize Emotions

The dataset has eight emotion labels that are named as follows: joy, sadness, fear, anger, surprise, neutral, disgust, and shame.

Natural Language Processing (NLP) gives machines the ability to break down and understand human language. Two of the top-used NLP techniques are Sentiment and Emotions analysis.

```python
finalwords =[]
for word in tokenizedwords:
    if word not in stopwords.words():#To remove stop words using the Gensim library, you can use the gensim. parsing. remove_stopwords method.
        finalwords.append(word)#/apend/ to add something to the end of a piece of writing:
emotion_list= []
with open('emotions.txt','r') as file:
    for line in file:
        clearline=line.replace('\n','').replace(',','').replace("'","").strip()#Python's strip() function can remove \n, a newline character, from a string.
        word,emotion=clearline.split(":")
        if word  in finalwords:
            emotion_list.append(emotion)
```

```python
print(emotion_list)
w=Counter(emotion_list)
```

#Sentiment_Analyse

Sentiment analysis, also referred to as opinion mining, is an approach to natural language processing (NLP) that identifies the emotional tone behind a body of text. This is a popular way for organizations to determine and categorize opinions about a product, service or idea.



```python
def sentiment_analyse(sentiment_text):
    score=SentimentIntensityAnalyzer().polarity_scores(sentiment_text)#Polarity refers to the overall sentiment conveyed by a particular text, phrase or word.
    neg=score['neg']
    pos=score['pos']
    if neg>pos:
        sentiment="Negative sentiment"
        print(sentiment)
    elif pos>neg:
        sentiment="positive sentiment"
        print(sentiment)
    else:
        sentiment="Neutral vibe"
        print(sentiment)
```

#Ploting Output

```python
sentiment_analyse(cleanedtext)#Simply put, sentiment analysis determines how the author feels about a certain topic.
fig, ax1=plt.subplots()#figure object to be used as a container for all the subplots. ax : A single object of the axes.
ax1.bar(w.keys(), w.values())#In seaborn barplot with bar, values can be plotted using sns. barplot() function and the sub-method containers returned by sns. barplot(). Import pandas, numpy, and seaborn
fig.autofmt_xdate()#The autofmt_xdate() method figure module of matplotlib library is used to rotate them and right align them.
plt.savefig("graph.png")
plt.show()
```
